```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(0)
x = np.linspace(-3,3,20).reshape(-1,1)
y = x ** 2 + np.random.normal(0,1, x.shape)
```

```python
model = LinearRegression().fit(x,y)
poly_feat = PolynomialFeatures(degree=3)
poly_model = LinearRegression().fit(poly_feat.fit_transform(x), y)
```

```python
y_linear_pred = model.predict(x)
y_poly_pred = poly_model.predict(poly_feat.transform(x))
```

```python
plt.scatter(x, y, color="black", label="Actual Data", s=20)
plt.plot(x, y_linear_pred, color="blue", linewidth=2)
plt.plot(x, y_poly_pred, color="red", linewidth=2)
plt.show()
```
